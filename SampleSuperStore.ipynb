```
df_num = df.select_dtypes("number")
df_num = df_num.drop(columns = "Postal Code")
df_num["State"] = df["State"]
```

Parece ser que los estados con una ganancia negativa son aquellos que ofrecen un mayor descuento en sus tiendas, esto puede ser debido directamente a las bajas ventas de un producto o simplemente por promocion.

```
mean_state = df_num.groupby("State").mean().sort_values(by = "Profit", ascending = False)
mean_state
```

Seleccionando unicamente aquellos estados en los que la ganancia es negativa podemos notar que efectivamente son los mismos que ofrecen un descuento

```
mean_low = mean_state.loc[mean_state["Profit"] <= 0].sort_values(by = "Profit")
mean_low
```

Y comparandolos con otros estados tienen unas ventas relativamente normales, inclusive estados con ventas menores no tienen perdidas

```
total_table = pd.pivot_table(df, values = "Quantity", index = "State", columns = "Category", aggfunc = np.mean)
total_table
```

Los estados con las menores ventas por categoria son: Main en muebles, Kansas en insumos de oficina y Connecticut en tecnologia, ninguno de ellos se encuentra en la lista de ganancia negativa

```
total_table.describe()
```

```
mean_low = mean_low.reset_index()
low_cities = df[df["State"].isin(mean_low["State"].tolist())].drop(columns = "Postal Code")
```

```
low_cities = low_cities.groupby("City").mean().sort_values(by = "Profit")
low_cities.shape
```

Tomando unicamente las ciudades con ganancia menor a cero vemos lo anterior, la media de las ventas se mantiene muy similar pero la media de los descuentos es de 0.39

```
low_cities = low_cities.loc[low_cities["Profit"] <= 0]
low_cities.describe()
```

A pesar de que muebles tiene una media de ventas considerable y que la cantidad esta en la media su media de ganancias es muy baja

```
categorias = df.drop(columns = ["Ship Mode", "Segment", "Country", "City", "State", "Postal Code", "Region"])
categorias = categorias.groupby("Category").mean()
categorias
```
